```python
def a(x):
  result=0
  while result<10:
    result=result+x
    if result==5:
      return result
  return result
b=a(1)
c=a(2)
print(b,c)
```

```python
def a(n):
  if n==0 or n==1:
    return 1
  else:
    return n+a(n-1)
print(a(3))
```
