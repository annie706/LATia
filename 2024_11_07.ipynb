```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

```python
df2=pd.DataFrame(np.random.randn(100,4),columns=list("ABCD"))
```

```python
x=df2['A']
y=df2['B']
#Add a constant to the model(for the intercept term)
x=sm.add_constant(x)
x
```

```python
#Fit the regression model
model=sm.OLS(y,x).fit()
#Display the regression results
print(model.summary())
```

```python
# Assuming df2 is your DataFrame with columns 'A', 'B', 'C', 'D'
import statsmodels.formula.api as smf

# Define the dependent variable and independent variables
y = df2['B']  # Dependent variable
x = df2[['A', 'C', 'D']]  # Independent variables

# Add a constant to the model (for the intercept term)
x = sm.add_constant(x)

# Fit the regression model
model = smf.ols('B ~ A + C + D', data=df2).fit()

# Display the regression results
print(model.summary())
```
