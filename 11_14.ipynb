```python
import scipy.stats as stats # import the stats module from scipy

group1 = [85, 86, 88, 75, 78, 94]
group2 = [92, 94, 89, 96, 93, 95, 76, 50]
group3 = [78, 81, 79, 84, 77, 80]
#  scipy.stats.f_oneway  ANOVA 
f_statistic, p_value = stats.f_oneway(group1, group2, group3) # Now stats is defined
# 
print("F :", f_statistic)
print("p-value:", p_value)
# 
alpha = 0.05 # 
if p_value < alpha:
  print("")
else:
  print("")
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
# 
data = {
'Score': [85, 86, 88, 75, 78, 94, 92, 94, 89, 96, 93, 95, 78, 81,
79, 84, 77, 80],
'Method': ['A', 'A', 'A', 'A', 'A', 'A', #  A
'B', 'B', 'B', 'B', 'B', 'B', #  B
'C', 'C', 'C', 'C', 'C', 'C'] #  C
}
#  DataFrame
df = pd.DataFrame(data)
#  statsmodels  ols  ANOVA 
model = ols('Score ~ C(Method)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
#  ANOVA 
print(anova_table)
```
